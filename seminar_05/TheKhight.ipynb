```python
BIG_NUM = 10 ** 6


class Khight():
    def __init__(self, size, idata, jdata):
        self.size = size
        self.xdata = idata
        self.ydata = jdata
        self.position = idata + size * (jdata - 1)

    def get_move_ability(self, position):
        ydata = position // self.size + 1
        xdata = position % self.size
        move_ability = [0 for _ in range(8)]
        if xdata + 2 in range(1, self.size + 1) and ydata + 1 in range(1, self.size + 1):
            move_ability[0] = 1
        if xdata + 1 in range(1, self.size + 1) and ydata + 2 in range(1, self.size + 1):
            move_ability[1] = 1
        if xdata - 1 in range(1, self.size + 1) and ydata + 2 in range(1, self.size + 1):
            move_ability[2] = 1
        if xdata - 2 in range(1, self.size + 1) and ydata + 1 in range(1, self.size + 1):
            move_ability[3] = 1
        if xdata - 2 in range(1, self.size + 1) and ydata - 1 in range(1, self.size + 1):
            move_ability[4] = 1
        if xdata - 1 in range(1, self.size + 1) and ydata - 2 in range(1, self.size + 1):
            move_ability[5] = 1
        if xdata + 1 in range(1, self.size + 1) and ydata - 2 in range(1, self.size + 1):
            move_ability[6] = 1
        if xdata + 2 in range(1, self.size + 1) and ydata - 1 in range(1, self.size + 1):
            move_ability[7] = 1
        return move_ability

    def move(self, direction, position):
        if direction == 0:
            return position + self.size + 2
        if direction == 1:
            return position + self.size * 2 + 1
        if direction == 2:
            return position + self.size * 2 - 1
        if direction == 3:
            return position + self.size - 2
        if direction == 4:
            return position - self.size - 2
        if direction == 5:
            return position - self.size * 2 - 1
        if direction == 6:
            return position - self.size * 2 + 1
        if direction == 7:
            return position - self.size + 2

    def bfs(self):
        vert_num = self.size ** 2
        dist = [BIG_NUM for _ in range(vert_num + 1)]
        dist[self.position] = 0
        queue = []
        queue.append(self.position)

        while queue:
            u = queue.pop(0)
            move_ability = self.get_move_ability(u)
            for v in range(8):
                if move_ability[v] == 1:
                    local_position = self.move(v, u)
                    # print(local_position)
                    if dist[local_position] == BIG_NUM:
                        dist[local_position] = dist[u] + 1
                        queue.append(local_position)

        return dist

    def find_shortest_way(self, xdata, ydata):
        final_position = self.size * (ydata - 1) + xdata
        return self.bfs()[final_position]

# МОЙ ПРИМЕР
size, i, j = 8, 1, 1
Horse = Khight(size, i, j)
x, y = 8, 8
print(f'Размер поля и координаты коня: {size}, {i, j}')
print(f'Конечные координаты коня: {x, y}')
print(f'Минимальное количество ходов: {Horse.find_shortest_way(x, y)}')
print()

size, i, j = list(map(int, input('Введите размер поля и координаты коня через пробел: ').split()))
Horse = Khight(size, i, j)
x, y = list(map(int, input('Введите конечные координаты через пробел: ').split()))
print(f'Минимальное количество ходов: {Horse.find_shortest_way(x, y)}')
```
